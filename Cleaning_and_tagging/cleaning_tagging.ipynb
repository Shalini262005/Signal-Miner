```python
import re

# Function to extract seniority level from job_title
def extract_seniority_from_title(title):
    title = title.lower()
    if re.search(r"\b(intern|trainee|entry level|junior)\b", title):
        return "Junior"
    elif re.search(r"\b(mid|associate|executive|assistant manager|team lead)\b", title):
        return "Mid"
    elif re.search(r"\b(senior|lead|manager)\b", title):
        return "Senior"
    elif re.search(r"\b(head|director|vp|vice president|chief|cto|ceo|coo)\b", title):
        return "Executive"
    else:
        return "Not Specified"

# Apply the extraction function to the job_title column
df["seniority"] = df["job_title"].apply(extract_seniority_from_title)
df.seniority.nunique()
```

```python
def assign_growth_signal(row):
    title = str(row['job_title']).lower()
    dept = str(row['department']).lower()
    seniority = str(row['seniority']).lower()
    
    # Rule-based tagging
    if 'product' in dept or 'product manager' in title:
        return "Product Launch"
    elif any(keyword in dept for keyword in ['sales', 'business development', 'marketing']) or \
         any(keyword in title for keyword in ['growth', 'gtm', 'go-to-market']):
        return "GTM Push"
    elif 'design' in dept or any(keyword in title for keyword in ['ux', 'ui', 'designer']):
        return "Design Overhaul"
    elif any(keyword in title for keyword in ['engineer', 'developer', 'data', 'machine learning', 'security', 'infra', 'platform']):
        return "Tech Modernization"
    elif any(keyword in title for keyword in ['compliance', 'risk', 'esg', 'sustainability']):
        return "Regulatory Focus"
    elif seniority in ['entry', 'mid']:
        return "Hiring Spree"
    else:
        return "General Growth"
```

```python
df['growth_signal_tag'] = df.apply(assign_growth_signal, axis=1)
df.head()
```

```python
department_counts = df.groupby(['company_name', 'department']).size().unstack(fill_value=0)
seniority_counts = df.groupby(['company_name', 'seniority']).size().unstack(fill_value=0)
job_counts = df.groupby('company_name').size().rename("total_jobs")

# Combine into one DataFrame
company_features = pd.concat([job_counts, department_counts, seniority_counts], axis=1).fillna(0)

# Ensure all expected columns exist
expected_dept_cols = ['Business Development', 'Marketing', 'Operations', 'Partnerships', 'Product', 'Sales', 'Tech', 'Other']
expected_seniority_cols = ['Executive', 'Junior', 'Mid', 'Senior', 'Not Specified']
for col in expected_dept_cols + expected_seniority_cols:
    if col not in company_features.columns:
        company_features[col] = 0

# Define tagging logic
def enriched_growth_tag(row):
    total_jobs = row['total_jobs']
    departments = row[expected_dept_cols]
    tech_roles = row['Tech']
    sales_roles = row['Sales']
    bd_roles = row['Business Development']
    marketing_roles = row['Marketing']
    diverse_depts = (departments > 0).sum()
    senior_or_mid_roles = row['Senior'] + row['Mid']

    if total_jobs > 15 and diverse_depts >= 3 and senior_or_mid_roles >= total_jobs * 0.5:
        return 'Hiring Spree'
    if (sales_roles + bd_roles + marketing_roles) >= total_jobs * 0.4:
        return 'GTM Push'
    if tech_roles >= total_jobs * 0.4:
        return 'Tech Modernization'
    if 5 <= total_jobs <= 15 and diverse_depts >= 2:
        return 'Multi-Signal Growth'
    return 'General Growth'

# Apply the tagging function
company_features['enhanced_growth_signal_tag'] = company_features.apply(enriched_growth_tag, axis=1)

# Reset index to make company_name a column
company_features = company_features.reset_index()

# View the results
company_features[['company_name', 'total_jobs', 'enhanced_growth_signal_tag']]
```

```python
company_features = company_features.sample(frac=1).reset_index(drop=True)
company_features
```
