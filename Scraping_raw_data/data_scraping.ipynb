```python
def scrape_razorpay():
    url = "https://job-boards.greenhouse.io/razorpaysoftwareprivatelimited"
    response = requests.get(url)
    soup = BeautifulSoup(response.text,"html.parser")
    
    jobs = []
    
    departments = soup.select("div.job-posts--table--department")
    for dept in departments:
        department = dept.find("h3",class_="section-header").text.strip()
        
        job_rows = dept.select("tbody tr.job-post")
        
        for job in job_rows:
            anchor = job.find("a")
            job_title = anchor.find("p",class_="body body--medium").text.strip()
            location = anchor.find("p",class_="body body__secondary body--metadata").text.strip()
            if location=='Bangalore':
                location='Bengaluru'
            else:
                location=location
            job_link = anchor["href"]
            
            jobs.append({
                "company_name": "Razorpay",
                "job_title": job_title,
                "department": department,
                "location": location,
                "job_posted_url": job_link,
                "scrape_date": datetime.today().strftime("%Y-%m-%d")
            })
    return jobs

razorpay_jobs = scrape_razorpay()
df1 = pd.DataFrame(razorpay_jobs)
df1.head()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.edge.service import Service as EdgeService
from selenium.webdriver.edge.options import Options
from datetime import datetime
import pandas as pd
import time
```

```python
def scrape_cred():
    options = Options()
    options.add_argument("--headless") 
    options.add_argument("--disable-gpu")
    options.add_argument("--window-size=1920,1080")

    service = EdgeService()

    driver = webdriver.Edge(service=service, options=options)
    url = "https://careers.cred.club/openings"
    driver.get(url)

    time.sleep(2)

    jobs = []
    job_urls = ['https://careers.cred.club/openings/c85d5832-53c2-413e-b711-41b5acef6a74','https://careers.cred.club/openings/ca6cd6ec-ff19-4a69-b03a-16b6ed0cbda8','https://careers.cred.club/openings/fad1748a-084f-4d0c-9ebb-aea8e0796392','https://careers.cred.club/openings/fd4bac8c-94ba-4549-bc10-41d942b0481c']
    job_cards = driver.find_elements(By.CSS_SELECTOR, "div.border")

    for (i,card) in enumerate(job_cards):
        try:
            department = card.find_element(By.CSS_SELECTOR, "div.uppercase").text.strip()
            department = capitalize_first_letter(department)
            title = card.find_element(By.CSS_SELECTOR, "div.lowercase").text.strip()
            details = card.find_elements(By.CSS_SELECTOR, "div.semi-bold-12")
            location = details[1].text.strip() if len(details) > 1 else ""
            location = location.split('\n')[2]
            location = capitalize_first_letter(location)

            try:
                job_link_element = card.find_element(By.CSS_SELECTOR, "div.cursor-pointer")
                job_link = job_link_element.get_attribute("onclick")
                job_url = url if not job_link else job_link
            except:
                job_url = url

            jobs.append({
                "company_name": "CRED",
                "job_title": title,
                "department": department,
                "location": location,
                "job_posted_url": job_urls[i-1],
                "scrape_date": datetime.today().strftime("%Y-%m-%d")
            })
        except Exception as e:
            print("Error parsing a job card:", e)

    driver.quit()

    return jobs
    
cred_jobs = scrape_cred()
df2 = pd.DataFrame(cred_jobs)
df2.head()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime
```

```python
def scrape_groww():

    url = 'https://job-boards.eu.greenhouse.io/groww'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    jobs = []

    for dept_header in soup.select('h3.section-header.font-primary'):
        department = dept_header.get_text(strip=True)
    
        job_table_div = dept_header.find_next_sibling('div', class_='job-posts--table')
    
        if not job_table_div:
            continue
    
        for job_row in job_table_div.select('tr.job-post'):
            title_tag = job_row.find('p', class_='body--medium')
            location_tag = job_row.find('p', class_='body body__secondary body--metadata').text.strip()
            if '-' in location_tag:
                location= location_tag.split('-')[0]
            elif '-' not in location_tag and ',' in location_tag:
                location = location_tag.split(',')[0]
            else:
                location_tag
            link_tag = job_row.find('a', href=True)

            if title_tag and location_tag and link_tag:
                jobs.append({
                    'company_name': 'Groww',
                    'job_title': title_tag.text.strip(),
                    'department': department,
                    'location': location,
                    'job_posted_url': link_tag['href'],
                    'scrape_date': datetime.today().strftime("%Y-%m-%d")
                })

    return jobs

groww_jobs = scrape_groww()
df3 = pd.DataFrame(groww_jobs)
df3.head()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import pandas as pd
import time
from datetime import datetime
```

```python
def scrape_meesho():

    options = webdriver.ChromeOptions()
    options.add_argument("--headless") 
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

    driver.get("https://meesho.io/jobs")
    time.sleep(5)

    soup = BeautifulSoup(driver.page_source, "html.parser")
    driver.quit()
    
    jobs = []
    
    job_section = soup.find("div", class_="open-positions")
    job_cards = job_section.find_all("div", class_="job")


    for job in job_cards:
        try:
            anchor = job.find("a")
            job_url = "https://meesho.io" + anchor["href"]

            title_div = anchor.find("div", class_=lambda c: c and "text-primary" in c)
            if not title_div:
                print("Title not found for:", job_url)
                continue
            title = title_div.text.strip()

            dept_div = anchor.find_all("div", class_=lambda c: c and "text-dark" in c)
            department = dept_div[1].text.strip() if len(dept_div) > 1 else "N/A"

            location_span = anchor.find("span", string=lambda text: text and ',' in text)
            location = location_span.text.strip() if location_span else "N/A"
            location = location.split(',')[0].strip()
            if location=='Bangalore':
                location='Bengaluru'
            else:
                location=location

            jobs.append({
                    'company_name': 'Meesho',
                    'job_title': title,
                    'department': department,
                    'location': location,
                    'job_posted_url': job_url,
                    'scrape_date': datetime.today().strftime("%Y-%m-%d")
                })
        except Exception as e:
            print(f"Error parsing job: {e}")
    return jobs
meesho_jobs = scrape_meesho()
df4 = pd.DataFrame(meesho_jobs)
df4.head()
```

Paytm

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from datetime import datetime
```

```python
def scrape_paytm():
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')
    options.add_argument('--disable-gpu')
    options.add_argument('--no-sandbox')

    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
    url = 'https://jobs.lever.co/paytm'
    driver.get(url)
    
    try:
        WebDriverWait(driver, 15).until(
            EC.presence_of_all_elements_located((By.CSS_SELECTOR, "div.postings-group"))
        )
    except Exception as e:
        driver.quit()
        return []

    jobs = []
    company_name = 'Paytm'
    scrape_date = datetime.now().strftime('%Y-%m-%d')

    try:
        job_sections = driver.find_elements(By.CSS_SELECTOR, "div.postings-group")

        for section in job_sections:
            try:
                department = section.find_element(By.CSS_SELECTOR, "div.large-category-header").text.strip()
                department = capitalize_first_letter(department)
                
                postings = section.find_elements(By.CSS_SELECTOR, "div.posting")

                for post in postings:
                    try:
                        job_title_element = post.find_element(By.CSS_SELECTOR, "a.posting-title")
                        title = job_title_element.find_element(By.CSS_SELECTOR, "h5[data-qa='posting-name']").text.strip()
                        job_url = job_title_element.get_attribute('href')

                        location_element = post.find_element(By.CSS_SELECTOR, "span.location")
                        location = location_element.text.strip()
                        location = capitalize_first_letter(location.split(',')[0])

                        jobs.append({
                            'company_name': company_name,
                            'job_title': title,
                            'department': department,
                            'location': location,
                            'job_posted_url': job_url,
                            'scrape_date': scrape_date
                        })

                    except Exception as e:
                        print(f"Error extracting job data: {e}")

            except Exception as e:
                print(f"Error extracting section data: {e}")

    except Exception as e:
        print(f"Error extracting job sections: {e}")

    driver.quit()
    return jobs

paytm_jobs = scrape_paytm()
df6 = pd.DataFrame(paytm_jobs)
df6.head()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
```

```python
def scrape_zolve():
    base_url = "https://zolve.freshteam.com"
    options = webdriver.ChromeOptions()
    options.add_argument("--headless")

    driver = webdriver.Chrome(service=Service(), options=options)
    driver.get("https://zolve.freshteam.com/jobs")
    time.sleep(3)

    jobs = []
    company_name = 'Zolve'
    scrape_date = datetime.today().strftime('%Y-%m-%d')

    sections = driver.find_elements(By.CSS_SELECTOR, 'li[data-portal-role^="_role_"]')

    for section in sections:
        try:
            department_full_text = section.find_element(By.CLASS_NAME, "role-title").text.strip()
            department = department_full_text.split("\n")[0].strip()

            job_links = section.find_elements(By.CSS_SELECTOR, 'a.heading.show')

            for job in job_links:
                try:
                    title = job.find_element(By.CLASS_NAME, "job-title").text.strip()
                    location_info = job.find_element(By.CLASS_NAME, "location-info").text.strip().split("\n")
                    location = location_info[0].strip()

                    relative_url = job.get_attribute("href")
                    url = relative_url if relative_url.startswith("http") else base_url + relative_url

                    jobs.append({
                        'company_name': company_name,
                        'job_title': title,
                        'department': department,
                        'location': location,
                        'job_posted_url': url,
                        'scrape_date': scrape_date
                    })
                except Exception as e:
                    print(f"Error parsing job card: {e}")
                    continue

        except Exception as e:
            print(f"Error processing section: {e}")
            continue

    driver.quit()
    return jobs

zolve_jobs = scrape_zolve()
df8 = pd.DataFrame(zolve_jobs)
df8.head()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from datetime import datetime
import time
import pandas as pd

def scrape_dripcapital():
    base_url = "https://dripcapital.hire.trakstar.com"
    careers_url = f"{base_url}/?q=&limit=100"

    options = webdriver.ChromeOptions()
    options.add_argument("--headless")
    driver = webdriver.Chrome(service=Service(), options=options)

    driver.get(careers_url)
    time.sleep(3)

    jobs = []
    company_name = 'Drip Capital'
    scrape_date = datetime.today().strftime('%Y-%m-%d')

    job_cards = driver.find_elements(By.CLASS_NAME, "js-card")

    for card in job_cards:
        try:
            title_elem = card.find_element(By.CLASS_NAME, "js-job-list-opening-name")
            job_title = title_elem.text.strip()

            job_url_rel = card.find_element(By.TAG_NAME, "a").get_attribute("href")
            job_posted_url = job_url_rel if job_url_rel.startswith("http") else base_url + job_url_rel

            city = card.find_element(By.CLASS_NAME, "meta-job-location-city").text.strip()
            state = card.find_element(By.CLASS_NAME, "meta-job-location-state").text.strip()
            country = card.find_element(By.CLASS_NAME, "meta-job-location-country").text.strip()
            location = f"{city}, {state}, {country}"

            rb_text_divs = card.find_elements(By.CLASS_NAME, "rb-text-4")

            department = ""
            for div in rb_text_divs:
                if "js-job-list-opening-meta" not in div.get_attribute("class"):
                    department = div.text.strip()
                    break 


            jobs.append({
                'company_name': company_name,
                'job_title': job_title,
                'location': location,
                'department': department,
                'job_posted_url': job_posted_url,
                'scrape_date': scrape_date
            })
        except Exception as e:
            print(f"Error: {e}")
            continue

    driver.quit()
    return jobs

drip_jobs = scrape_dripcapital()
df10 = pd.DataFrame(drip_jobs)
df10.head()
```

```python
import re

# Define function to extract seniority from job titles
def extract_seniority(title):
    title = title.lower()
    if any(kw in title for kw in ['intern']):
        return 'Intern'
    elif any(kw in title for kw in ['junior', 'entry']):
        return 'Junior'
    elif any(kw in title for kw in ['mid', 'mid-level', 'associate']):
        return 'Mid'
    elif any(kw in title for kw in ['senior', 'sr']):
        return 'Senior'
    elif any(kw in title for kw in ['lead', 'principal']):
        return 'Lead'
    elif any(kw in title for kw in ['director', 'head']):
        return 'Director'
    else:
        return 'Other'

# Normalize location (basic logic for now)
def normalize_location(loc):
    loc_lower = loc.lower()
    if 'remote' in loc_lower:
        return 'Remote'
    elif 'hybrid' in loc_lower:
        return 'Hybrid'
    else:
        return loc.strip()

# Create the transformed DataFrame
transformed_df = pd.DataFrame({
    'company_name': df['company_name'],
    'job_title': df['job_title'],
    'department': df['department'],  # Assumes already mapped
    'location': df['location'].apply(normalize_location),
    'seniority': df['job_title'].apply(extract_seniority),
    'job_url': df['job_posted_url'],
    'growth_signal_tag': ''  # Placeholder for future tagging
})

transformed_df.head()
```
